## Bar Chart

```python
# Create the bar chart again with title and diagonal x-axis labels
fig, ax = plt.subplots(figsize=(12, 6))
df_age.plot.bar(x='Age Group', y='Population', title='Age Population in Los Angeles County', ax=ax, color='green', legend=False)
# Rotate x-axis labels diagonally
ax.set_xticklabels(df_age['Age Group'], rotation=45, ha='right')
```

```python
# based on having 12 age categories, I will merge ages to have four categories
# Define new age categories and sum populations accordingly
df_age_grouped = pd.DataFrame({
    'Age Group': [
        'Children & Adolescents (0-17)', 
        'Young Adults (18-34)', 
        'Middle-Aged Adults (35-64)', 
        'Older Adults (65 and Over)'
    ],
    'Population': [
        df_age.loc[df_age['Age Group'].isin([
            'Under 5 Years', '5 to 9 Years', '10 to 14 Years', '15 to 17 Years'
        ]), 'Population'].sum(),
        
        df_age.loc[df_age['Age Group'].isin([
            '18 to 24 Years', '25 to 34 Years'
        ]), 'Population'].sum(),
        
        df_age.loc[df_age['Age Group'].isin([
            '35 to 44 Years', '45 to 54 Years', '55 to 64 Years'
        ]), 'Population'].sum(),
        
        df_age.loc[df_age['Age Group'].isin([
            '65 to 74 Years', '75 to 84 Years', '85 Years and Over'
        ]), 'Population'].sum()
    ]
})
```

```python
# Create a bar chart with merged age categories
fig, ax = plt.subplots(figsize=(10, 6))
df_age_grouped.plot.bar(x='Age Group', y='Population', ax=ax, color='blue', legend=False)

# Customize title and labels
ax.set_title('Total Population by Age Group in Los Angeles County 2023', fontsize=14)
ax.set_xlabel('Age Group', fontsize=12)
ax.set_ylabel('Total Population', fontsize=12)

# Rotate x-axis labels diagonally for better readability
ax.set_xticklabels(df_age_grouped['Age Group'], rotation=45, ha='right')

# Set y-axis ticks dynamically
yticks = range(0, max(df_age_grouped['Population']) + 100000, 100000)
ax.set_yticks(yticks)
ax.set_yticklabels([f'{int(tick)}' for tick in yticks])

# Display the chart
plt.show()
```

## Histogram

```python
print(df2.columns.to_list())
```

```python
# Update age categories with correct column names based on df2
age_categories = {
    'Children & Adolescents (0-17)': ['Under 5 Years', '5 to 9 Years', '10 to 14 Years', '15 to 17 Years'],
    'Young Adults (18-34)': ['18 to 24 Years', '25 to 34 Years'],
    'Middle-Aged Adults (35-64)': ['35 to 44 Years', '45 to 54 Years', '55 to 64 Years'],
    'Older Adults (65 and Over)': ['65 to 74 Years', '75 to 84 Years', '85 Years and Over']
}

# Create a new dataframe with aggregated categories
df_grouped = pd.DataFrame()

for group, columns in age_categories.items():
    existing_columns = [col for col in columns if col in df2.columns]  # Ensure columns exist
    df_grouped[group] = df2[existing_columns].sum(axis=1)  # Sum relevant columns

# Display the first few rows to confirm
import matplotlib.pyplot as plt

# Plot histograms for each age category
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # 2x2 grid for 4 categories
axes = axes.flatten()  # Flatten the axes array for easy iteration

for idx, (group, _) in enumerate(age_categories.items()):
    df_grouped[group].plot.hist(ax=axes[idx], bins=30, alpha=0.7, color='blue', edgecolor='black')
    axes[idx].set_title(f'Distribution of {group}')
    axes[idx].set_xlabel('Population')
    axes[idx].set_ylabel('Frequency')

plt.tight_layout()  # Adjust layout for better readability
plt.show()
```

## Maps

```python
# introduce tools to map
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Suppress warnings related to chained assignments
pd.options.mode.chained_assignment = None
```

```python
# Load LA County census tracts shapefile
gdf = gpd.read_file("data_census/Census_Tracts_2020.geojson")
```

```python
print(gdf.columns)
```

```python
# Convert to string for matching
df2["Geo_TRACT"] = df2["Geo_TRACT"].astype(str)
gdf["CT20"] = gdf["CT20"].astype(str)

# Merge using CT20
gdf_merged = gdf.merge(df2, left_on="CT20", right_on="Geo_TRACT", how="left")

# Check if merge
print(gdf_merged.head())
```

```python
# Define new age groups by summing relevant columns
df2["Children & Adolescents (0-17)"] = df2[['Under 5 Years', '5 to 9 Years', '10 to 14 Years', '15 to 17 Years']].sum(axis=1)
df2["Young Adults (18-34)"] = df2[['18 to 24 Years', '25 to 34 Years']].sum(axis=1)
df2["Middle-Aged Adults (35-64)"] = df2[['35 to 44 Years', '45 to 54 Years', '55 to 64 Years']].sum(axis=1)
df2["Older Adults (65 and Over)"] = df2[['65 to 74 Years', '75 to 84 Years', '85 Years and Over']].sum(axis=1)

# Merge the grouped data with the GeoJSON
gdf_merged = gdf.merge(df2, left_on="CT20", right_on="Geo_TRACT", how="left")

# Check if the column exists now
print(gdf_merged.columns)
```

```python
# Select only necessary columns
columns_to_keep = [
    "CT20", "geometry",  # Keep Census Tract ID & Geometry
    "Total Population", 
    "Children & Adolescents (0-17)",
    "Young Adults (18-34)", 
    "Middle-Aged Adults (35-64)", 
    "Older Adults (65 and Over)"
]
```

```python
# Create a clean GeoDataFrame
gdf_cleaned = gdf_merged[columns_to_keep]
```

```python
# Display the first few rows
print(gdf_cleaned.head())
```

```python
# Create map with clean data for Children and Adolescents
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

gdf_cleaned.plot(column="Children & Adolescents (0-17)",  
                cmap="Blues",  
                linewidth=0.5, 
                edgecolor="black",
                legend=True,
                ax=ax)

ax.set_title("Children & Adolescents (0-17) by Census Tract in Los Angeles County", fontsize=14)
plt.axis("off")  # Remove x and y axis labels

plt.show()
```

```python
# Create map with clean data for Young Adults (18-34)
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

gdf_cleaned.plot(column="Young Adults (18-34)",  
                cmap="Blues",  
                linewidth=0.5, 
                edgecolor="black",
                legend=True,
                ax=ax)
# Remove x and y axis labels
ax.set_title("Young Adults (18-34) by Census Tract in Los Angeles County", fontsize=14)
plt.axis("off")  

plt.show()
```

```python
# Create map with clean data for Young Adults (18-34)
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

gdf_cleaned.plot(column="Middle-Aged Adults (35-64)",  
                cmap="Blues",  
                linewidth=0.5, 
                edgecolor="black",
                legend=True,
                ax=ax)
# Remove x and y axis labels
ax.set_title("Young Adults (18-34) by Census Tract in Los Angeles County", fontsize=14)
plt.axis("off")  

plt.show()
```

```python
# Create map with clean data for Young Adults (18-34)
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

gdf_cleaned.plot(column="Older Adults (65 and Over)",  
                cmap="Blues",  
                linewidth=0.5, 
                edgecolor="black",
                legend=True,
                ax=ax)
# Remove x and y axis labels
ax.set_title("Older Adults (65 and Over) by Census Tract in Los Angeles County", fontsize=14)
plt.axis("off")  

plt.show()
```
